```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Exponential growth $$b*e^{ax}$$

```python
x = np.arange(1, 100)
a = 0.1
b = 1
y= b*np.exp(a*(x))

plt.plot(x,y)
plt.show()
```

### Logistic growth $$\frac{K}{(1+e^{(x-b)-r})}$$


```python
x = np.arange(0, 100)
K = 100
b = 50
r = 0.1

y = K/(1+np.exp((x-b)*-r))

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 100)
K = 100
b = 50
r = 0.3
K2 = 400
b2 = 80
r2 = 0.3
```

```python
y = K/(1+np.exp((x-b)*-r)) + K2/(1+np.exp((x-b2)*-r2))
```
