```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Exponential growth $$βe^{αx}$$

```python
x = np.arange(1, 100)
a = 0.1
b = 1
y= b*np.exp(a*(x))

plt.plot(x,y)
plt.show()
```

### Logistic growth $$\frac{K}{(1+e^{(x-b)-r})}$$


```python
x = np.arange(0, 100)
K = 100
b = 50
r = 0.1
```

```python
y = K/(1+np.exp((x-b)*-r))
```

```python
plt.plot(x,y)
plt.show()
```
