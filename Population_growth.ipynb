```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```


### Exponential growth modell $$\frac{dP(t)}{dt}= aP(t)$$
#### OR
### $$P'(t)= aP(t)$$

### Exponential growth function $$P(t) = b\cdot e^{at}$$

```python
t = np.arange(1, 100)
a = 0.1
b = 1
y= b*np.exp(a*(t))

plt.plot(t,y)
plt.show()
```

### Logistic differential equation $$\frac{dP(t)}{dt} = a\cdot P(t)\cdot (1-\frac{P(t)}{K})$$
#### or
### $$P'(t)= a\cdot P(t)\cdot (1-\frac{P(t)}{K})$$

### Logistic growth function$$P(t)=\frac{K}{(1+e^{-a(t-b)})}$$


```python
t = np.arange(1, 100)
K = 100
a = 0.1
b = 50

y = K/(1+np.exp(-a*(t-b)))
plt.plot(t,y)
plt.xticks(np.linspace(10,100,10))
plt.show()
```

#### Calculating time required Δt for the curve to grow from 10% to 90%

$$\frac{9}{10}\cdot K = \frac{K}{1+e^{-a(t-b)}}$$
$$$$
$$ 9 = e^{-a(t-b)}$$
$$$$
$$ln(9) =a(t-b)$$
$$$$
$$\frac{ln(9)}{a}+b = t_{90}$$
$$$$


$$t_{90} - t_{10} = Δt$$
$$$$
$$(\frac{ln(9)}{a}+b) - (\frac{-ln(9)}{a}+b)= Δt$$
$$$$
$$(\frac{ln(9)+ln(9)}{a}= Δt$$
$$$$
$$(\frac{ln(81)}{Δt})= a$$

$$N(t) = \frac{K}{1+e^{\frac{-ln(81)}{Δt}(t-b)}}$$

### Bi-logistic growth

### $$N_{1}(t) = N_{1}(t)+N_{2}(t)$$$$ $$$$N(t_1) = \frac{K_1}{1+e^{\frac{-ln(81)}{Δt_1}(t-b_1)}}$$ $$$$ $$N(t_1) = \frac{K_2}{1+e^{\frac{-ln(81)}{Δt_2}(t-b_2)}}$$ 

### $$N(t) = \frac{K_1}{1+e^{\frac{-ln(81)}{Δt_1}(t-b_1)}} + \frac{K_2}{1+e^{\frac{-ln(81)}{Δt_2}(t-b_2)}}$$

```python
t = np.arange(1, 100)

K1 = 100
dt1 = 40
b1 = 50

K2 = 400
dt2 = 20 
b2 = 80

y = K1/(1+np.exp(-np.log2(81)/dt1*(t-b1))) + \
    K2/(1+np.exp(-np.log2(81)/dt2*(t-b2)))

plt.plot(t,y)
plt.show()
```

```python
y1 = K1/(1+np.exp(-np.log2(81)/dt1*(t-b1)))
y2 = K2/(1+np.exp(-np.log2(81)/dt2*(t-b2)))

plt.plot(t,y1)
plt.plot(t,y2)
plt.show()
```
